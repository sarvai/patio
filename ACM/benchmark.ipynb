```python
def imshow( im ):
    fig = pp.figure( figsize=(13,13) )
    pp.imshow( im[:,:,::-1] )
    
def draw_rects( im, rects, color ):
    for r in rects :
        cv2.rectangle( im, (r[0],r[1]), (r[2],r[3]), color[::-1],2 )
```
